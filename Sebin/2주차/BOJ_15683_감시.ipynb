```python
import copy
import sys

# 사각지대 영역 개수
def count_zero(graph):
    all_count = 0
    for i in graph:
        all_count += i.count(0)
    return all_count

# CCTV 방향
def cctv(depth,lis):
    global min_val
    
    graph = copy.deepcopy(lis)
    
    if depth==len_loca:
        min_val = min(count_zero(graph),min_val)
        return
        
    i = cctv_loca[depth][0]
    j = cctv_loca[depth][1]
    
    for p in dirc[li[i][j]-1]:
        cctv(depth + 1,see(i,j,p,graph))
        graph = copy.deepcopy(lis)

# 감시영역 체크
def see(r,c,p,graph):
    for i in p:
        ny = r
        nx = c
        while True:
            nx += dx[i]
            ny += dy[i]
            if nx >= m or ny >= n or nx < 0 or ny < 0 or li[ny][nx] ==6:
                break
            elif li[ny][nx]==0:
                graph[ny][nx]='#'  
    return graph
            
# 상하좌우 움직일 좌표
dx = [0,1,0,-1]
dy = [-1,0,1,0]

# CCTV별 바라보는 곳
dirc = [[[0],[1],[2],[3]]  ,[[0,2],[1,3]]  ,[[0,1],[1,2],[2,3],[3,0]]  ,[[3,0,1],[0,1,2],[1,2,3],[2,3,0]]  ,[[0,1,2,3]]]

n,m = map(int,sys.stdin.readline().split())
li=[]
cctv_loca=[]
len_loca=0

for i in range(n):
    l = list(map(int,sys.stdin.readline().split()))
    li.append(l)
    for j in range(m): # CCTV 위치
        if 0 < l[j] < 6:
            len_loca +=1
            cctv_loca.append([i,j])

min_val = 1e9
cctv(0,li)
print(min_val)
```
